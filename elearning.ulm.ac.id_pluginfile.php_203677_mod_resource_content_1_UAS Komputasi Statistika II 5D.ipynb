**Gambarkan distribusi data untuk variabel Usia, Penghasilan, dan Limit Kredit.**
<Br> **Berikan warna berbeda untuk setiap variabel!**
<Br> **Berilah judul yang sesuai untuk setiap gambarnya!**

```python
#USIA
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='USIA', color='blue')
plt.title('Distribusi Data Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.show()

#PENGHASILAN
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='PENGHASILAN', color='green')
plt.title('Distribusi Data Penghasilan')
plt.xlabel('Penghasilan')
plt.ylabel('Frekuensi')
plt.show()

#LIMIT KKREDIT
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='LIMIT KREDIT', color='red')
plt.title('Distribusi Data Limit Kredit')
plt.xlabel('Limit Kredit')
plt.ylabel('Frekuensi')
plt.show()
```

**Tunjukkan Persentase/Proporsi Status Upselling *TOP* dan *NEW*.**

```python
sns.barplot(x='STATUS UPSELLING', y='PENGHASILAN', data=data, ci=None)
plt.title('Hubungan Antara Status Upselling dengan Penghasilan')
plt.xlabel('Status Upselling')
plt.ylabel('Penghasilan')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie( data['STATUS UPSELLING'].value_counts().values, labels= data['STATUS UPSELLING'].value_counts().index, autopct='%1.1f%%')
ax.set_title("Distribusi Data", fontsize=18, fontweight='bold', color='r')

plt.show()
```

**Berapakah Rata-rata Limit Kredit Nasabah dengan Status Upselling *TOP* dan *NEW***

```python
print("Rata-rata Limit Kredit Nasabah dengan Status Upselling TOP:", np.mean(data[data['STATUS UPSELLING'] == 'TOP']['LIMIT KREDIT']))
print("Rata-rata Limit Kredit Nasabah dengan Status Upselling NEW:", np.mean(data[data['STATUS UPSELLING'] == 'NEW']['LIMIT KREDIT']))
```

**Berapakah Median Usia Nasabah dengan Status Upselling *TOP* dan *NEW*?**

```python
print("Median Limit Kredit Nasabah dengan Status Upselling TOP:", np.median(data[data['STATUS UPSELLING'] == 'TOP']['USIA']))
print("Median Limit Kredit Nasabah dengan Status Upselling NEW:", np.median(data[data['STATUS UPSELLING'] == 'NEW']['USIA']))
```

**Tampilkan frekuensi Jenis Pekerjaan terhadap Status Upselling.**

```python
sns.countplot(x="JENIS PEKERJAAN", data=data, hue="STATUS UPSELLING", palette="Dark2")
plt.title('Frekuensi Jenis Pekerjaan terhadap Status Upselling')
plt.xlabel('Jenis Pekerjaan')
plt.ylabel('Frekuensi')
plt.show()
```

**Periksalah Menggunakan Grafik Apakah Ada Hubungan Antara Usia dengan Penghasilan.**
<br>**Berilah interpretasi dari hasil grafiknya.**

```python
sns.relplot(x="USIA", y="PENGHASILAN", data=data, kind="scatter")
plt.title('Hubungan Antara Usia dengan Penghasilan')
plt.xlabel('Usia')
plt.ylabel('Penghasilan')
plt.show()
```

Berdasarkan scatterplot diatas tidak tampak hubungan jelas antara variabel usia dan penghasilan. Terlihat oleh pola data yang terbentuk, tidak berbentuk pola tertentu seperti garis linier atau lainnya.

**Periksalah Menggunakan Grafik Apakah Ada Hubungan Antara Status Upselling dengan Penghasilan.**
<br>**Berilah interpretasi dari hasil grafiknya.**

```python
sns.barplot(x='STATUS UPSELLING', y='PENGHASILAN', data=data, ci=None)
plt.title('Hubungan Antara Status Upselling dengan Penghasilan')
plt.xlabel('Status Upselling')
plt.ylabel('Penghasilan')
plt.show()
```

Berdasarkan Diagram Batang diatas, terlihat bahwa seseorang dengan status upselling `NEW` cenderung memiliki penghasilan yang lebih tinggi jika dibandingkan dengan seseorang yang memiliki status upselling `TOP`. 

**Periksalah Menggunakan Grafik Apakah Ada Hubungan Antara Status Tempat Tinggal dengan Penghasilan.**
<br>**Berilah interpretasi dari hasil grafiknya.**
<br>**Urutkanlah grafiknya dari penghasilan terendah ke tertinggi .**

```python
sns.barplot(x='STATUS TEMPAT TINGGAL', y='PENGHASILAN', data=data, ci=None)
plt.title('Hubungan Antara Status Tempat Tinggal dengan Penghasilan')
plt.xlabel('Status Tempat Tinggal')
plt.ylabel('Penghasilan')
plt.show()
```

Berdasarkan hasil visualisasi menggunakan diagram batang diatas, didapatkan bahwa seseorang dengan status tempat tinggal merupakan `rumah dinas` memiliki penghasilan yang cenderung lebih tinggi sedangkan seseorang dengan status tempat tinggal `sewa` atau status tempat tinggal `Milik Orang Tua`  cenderung memiliki penghasilan yang lebih kecil. Hal ini berbanding lurus dengan kenyataan kebanyakn pada saat ini dimana biasanya para pejabat/ pemangku kepentingan suatu perusahaan atau kantor atau bahkan petinggi negara itu menempati rumah dinas sedangkan seseorang yang memiliki penghasilan yang lebih kecil cenderung akan tinggal bersama orang tua atau memilih menyewa tempat tinggal.

```python
sns.barplot(x='STATUS TEMPAT TINGGAL', y='PENGHASILAN', data=data, ci=None, order=data.groupby('STATUS TEMPAT TINGGAL')['PENGHASILAN'].mean().sort_values().index)
plt.title('Hubungan Antara Status Tempat Tinggal dengan Penghasilan')
plt.xlabel('Status Tempat Tinggal')
plt.ylabel('Penghasilan')
plt.show()
```
